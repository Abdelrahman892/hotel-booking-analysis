##### Number of bookings by hotel type


```python
number_of_bookings_by_hotel = df["hotel"].value_counts().reset_index()
number_of_bookings_by_hotel.columns = ["hotel", "bookings"]
print(f"number of bookings by Hotel type {number_of_bookings_by_hotel}")

sns.barplot(data=number_of_bookings_by_hotel, x="hotel", y="bookings")


plt.title("number of bookings by Hotel type")
plt.show()
```

#### Insight: City Hotel has significantly more bookings than Resort Hotel

##### Cancellation Rate for Each Hotel Type

```python
Cancellation_Rate_for_Each_Hotel = (
    df.groupby("hotel")["is_canceled"].mean().reset_index()
)
Cancellation_Rate_for_Each_Hotel.columns = ["hotel", "cancellation_rate"]

print(f"Cancellation Rate for Each Hotel Type {Cancellation_Rate_for_Each_Hotel}")
```

```python
sns.barplot(data=Cancellation_Rate_for_Each_Hotel, x="hotel", y="cancellation_rate")

plt.title("Cancellation Rate by Hotel Type")
plt.ylabel("Cancellation Rate")
plt.show()
```

### insight: City Hotel dominates in booking volume, however it also has a higher cancellation rate, indicating that city bookings may be more flexible or driven by short-term demand compared to Resort Hotel bookings.

#### Number of bookings by country


```python
number_of_bookings_by_country = df["country"].value_counts().reset_index()
number_of_bookings_by_country.columns = ["country", "bookings"]
print(f"number of bookings by country: {number_of_bookings_by_country}")
```

```python
top10 = number_of_bookings_by_country.head(10)

plt.figure(figsize=(10, 6))

sns.barplot(
    data=top10,
    x="bookings",
    y="country",
    orient="h"
)

plt.title("Top 10 Countries by Number of Bookings")
plt.xlabel("Number of Bookings")
plt.ylabel("Country")

# values on bars
for i, v in enumerate(top10["bookings"]):
    plt.text(v, i, str(v), va='center')

plt.tight_layout()
plt.show()
```

##### insight: Bookings are highly concentrated in a few countries, indicating that demand is dominated by a small set of international markets rather than being evenly distributed.

##### Does lead time affect cancellation?

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    data=df,
    x="is_canceled",
    y="lead_time",
    palette="Set2"
)

plt.title("Lead Time by Cancellation Status")
plt.xlabel("Is Canceled")
plt.ylabel("Lead Time")

plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

#### The analysis shows that longer lead times are associated with a higher likelihood of cancellation. This suggests that bookings made well in advance carry more uncertainty and are more prone to changes in customer plans.

##### Analysis of Repeat Guest Behavior in Hotel Bookings

```python
Repeated_guest = df["is_repeated_guest"].value_counts().reset_index()
Repeated_guest.columns = ["is_repeated_guest", "count"]

Repeated_guest["is_repeated_guest"] = Repeated_guest["is_repeated_guest"].map(
    {0: "Non-Repeat", 1: "Repeat"}
)

plt.figure(figsize=(7, 5))

sns.barplot(
    data=Repeated_guest,
    x="is_repeated_guest",
    y="count",
    palette="Set2"
)

plt.title("Number of Repeat vs Non-Repeat Guests")
plt.xlabel("")
plt.ylabel("Count")

# values on bars
for i, v in enumerate(Repeated_guest["count"]):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

#### The majority of customers are not repeat guests, indicating low customer retention or low revisit rate

##### peak season

```python
month_order = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

peak_season = df["arrival_date_month"].value_counts().reset_index()
peak_season.columns = ["Month", "Number of bookings"]

peak_season["Month"] = pd.Categorical(
    peak_season["Month"],
    categories=month_order,
    ordered=True
)

peak_season = peak_season.sort_values("Month")

plt.figure(figsize=(10, 5))

sns.barplot(
    data=peak_season,
    x="Month",
    y="Number of bookings",
    palette="viridis"
)

plt.title("Peak Seasons")
plt.xlabel("Month")
plt.ylabel("Number of bookings")

plt.xticks(rotation=45)

# values on bars
for i, v in enumerate(peak_season["Number of bookings"]):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

#### August show the highest booking demand, indicating a strong seasonal peak during summer months

##### What is the relationship between ADR and reservation status?

```python
adr_reservation_status = df.groupby("reservation_status")["adr"].mean().reset_index()
adr_reservation_status.columns = ["reservation_status", "avg_adr"]

plt.figure(figsize=(8, 5))

sns.barplot(
    data=adr_reservation_status,
    x="reservation_status",
    y="avg_adr",
    palette="Set2"
)

plt.title("Average ADR by Reservation Status")
plt.xlabel("Reservation Status")
plt.ylabel("Average ADR")

# values on bars
for i, v in enumerate(adr_reservation_status["avg_adr"]):
    plt.text(i, v, f"{v:.2f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()

adr_reservation_status
```

#### Canceled reservations tend to have a higher average ADR compared to completed stays, suggesting that higher prices may be associated with increased booking cancellations.

##### What is the relationship between cancellation and deposit

```python
deposit_analysis = (
    df.groupby("deposit_type")
    .agg(
        total_bookings=("is_canceled", "count"),
        cancellations=("is_canceled", "sum"),
        cancellation_rate=("is_canceled", "mean"),
    )
    .reset_index()
)

plt.figure(figsize=(8, 5))

sns.barplot(
    data=deposit_analysis,
    x="deposit_type",
    y="cancellation_rate",
    palette="Set2"
)

plt.title("Cancellation Rate by Deposit Type")
plt.xlabel("Deposit Type")
plt.ylabel("Cancellation Rate")

plt.ylim(0, 1)  

# values on bars
for i, v in enumerate(deposit_analysis["cancellation_rate"]):
    plt.text(i, v, f"{v:.2f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()

deposit_analysis
```

##### Deposit type is a strong predictor of cancellation behavior. Surprisingly, non-refundable bookings show extremely high cancellation rates, indicating potential data or operational anomalies that require further investigation.

##### Which channel has the highest cancellation rate?

```python
channel_cancellation = (
    df.groupby("distribution_channel")
    .agg(
        total_bookings=("is_canceled", "count"),
        cancellations=("is_canceled", "sum"),
        cancellation_rate=("is_canceled", "mean"),
    )
    .reset_index()
)

plt.figure(figsize=(8, 5))

sns.barplot(
    data=channel_cancellation,
    x="distribution_channel",
    y="cancellation_rate",
    palette="Set2"
)

plt.title("Cancellation Rate by Distribution Channel")
plt.xlabel("Distribution Channel")
plt.ylabel("Cancellation Rate")

plt.ylim(0, 1)

# values on bars
for i, v in enumerate(channel_cancellation["cancellation_rate"]):
    plt.text(i, v, f"{v:.2f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()

channel_cancellation
```

#### Distribution channel significantly impacts cancellation behavior. TA/TO bookings contribute the majority of cancellations despite being the largest source of bookings, while direct bookings show the most reliable customer behavior.

```python
pivot_df = df.pivot_table(
    values="is_canceled",
    index="distribution_channel",
    columns="deposit_type",
    aggfunc="mean",
)

plt.figure(figsize=(8, 5))

sns.heatmap(
    pivot_df,
    annot=True,
    fmt=".2f",
    cmap="YlOrRd",   
    linewidths=0.5,
    linecolor="white"
)

plt.title("Cancellation Rate Heatmap", fontsize=14)
plt.xlabel("Deposit Type")
plt.ylabel("Distribution Channel")

plt.tight_layout()
plt.show()
```

##### What is the relationship between booking changes and cancellation status?

```python
booking_changes_cancellation_status = (
    df.groupby("booking_changes")["is_canceled"]
    .mean()
    .reset_index()
    .sort_values("booking_changes")
)

plt.style.use("dark_background")

plt.figure(figsize=(10, 6))

plt.plot(
    booking_changes_cancellation_status["booking_changes"],
    booking_changes_cancellation_status["is_canceled"],
    marker="o",
    linewidth=2,
    color="white"
)

plt.title("Cancellation Rate vs Booking Changes")
plt.xlabel("Booking Changes")
plt.ylabel("Cancellation Rate")

plt.grid(True, linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()
```

#### Cancellation rate tends to decrease as booking changes increase, suggesting that customers who modify their bookings are more committed and less likely to cancel. However, high-change groups have low sample sizes, which introduces variability in the trend.

##### How does Total special requests affect cancellation rate?

```python
total_special_requests_affect_cancellation_rate = (
    df.groupby("total_of_special_requests")["is_canceled"]
    .mean()
    .reset_index()
    .sort_values("total_of_special_requests")
)

plt.style.use("dark_background")

plt.figure(figsize=(10, 6))

plt.plot(
    total_special_requests_affect_cancellation_rate["total_of_special_requests"],
    total_special_requests_affect_cancellation_rate["is_canceled"],
    marker="o",
    linewidth=2,
    color="white"
)

plt.title("Cancellation Rate vs Special Requests")
plt.xlabel("Total Special Requests")
plt.ylabel("Cancellation Rate")

plt.grid(True, linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()
```

#### As the number of special requests increases, the cancellation rate tends to decrease.
